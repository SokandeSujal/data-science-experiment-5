```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
```

```python
# Load the data into a pandas dataframe
iris_data = pd.read_csv(url, header=None, names=columns)
```

```python
iris_data
```

```python
# Plot Sepal Length vs Sepal Width
plt.figure(figsize=(10, 6))
for species in iris_data['class'].unique():
    subset = iris_data[iris_data['class'] == species]
    plt.scatter(subset['sepal_length'], subset['sepal_width'], label=species)

plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend()
plt.show()
```

```python
# Plot Petal Length vs Petal Width
plt.figure(figsize=(10, 6))
for species in iris_data['class'].unique():
    subset = iris_data[iris_data['class'] == species]
    plt.scatter(subset['petal_length'], subset['petal_width'], label=species)

plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.legend()
plt.show()
```
